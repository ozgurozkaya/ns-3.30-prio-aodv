```python
import pandas
import glob
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pandas.read_csv('/home/ozgur/Desktop/PRIO-AODV/stats-module/VanetRCAODV-report-statsModule.csv', header=None)
df2 = pandas.read_csv('/home/ozgur/Desktop/PRIO-AODV/stats-module/VanetRCDSR-report-statsModule.csv', header=None)
df3 = pandas.read_csv('/home/ozgur/Desktop/PRIO-AODV/stats-module/PRIOAODV-report-statsModule.csv', header=None)
df4 = pandas.read_csv('/home/ozgur/Desktop/PRIO-AODV/stats-module/PRIOAODVEmergency-report-statsModule.csv', header=None)
df5 = pandas.read_csv('/home/ozgur/Desktop/PRIO-AODV/stats-module/VanetRCDSR-report-statsModule.csv', header=None)
connections = [50, 75, 100, 125]
pdfAODV = []
delayAODV = []
pdfDSR = []
delayDSR = []
pdfPRIOAODV = []
delayPRIOAODV = []
pdfPRIOAODV_EM = []
delayPRIOAODV_EM = []
pdfPRIOAODV_EM_raw = np.empty(2)
delayPRIOAODV_EM_raw = np.empty(2)

pdfPRIOAODV_EM_test = np.empty(2)
delayPRIOAODV_EM_test = np.empty(2)
'''
pdfOurs = [90.28, 97.18, 98.54, 98.11]
RXbitrateOurs = [756.09, 1374.43, 1967.31, 2699.27]

pdfTheirs = [98.41, 98.23, 99.26, 89.54]
RXbitrateTheirs = [737.34, 1473.81, 2203.75,  2671.18]
'''

df = df.dropna()
df2 = df2.dropna()
df3 = df3.dropna()
df4 = df4.dropna()
df5 = df5.fillna(0)
for x in range(1,5):
    
    tmp = df.loc[df[5] == 25+(25*x)]
    pdfAODV.append(tmp[2].mean())
    delayAODV.append(tmp[4].mean())
    
    tmp = df2.loc[df2[5] == 25+(25*x)]
    pdfDSR.append(tmp[2].mean())
    delayDSR.append(tmp[4].mean())
    
    tmp = df3.loc[df3[5] == 25+(25*x)]
    pdfPRIOAODV.append(tmp[2].mean())
    delayPRIOAODV.append(tmp[4].mean())
    
    tmp = df4.loc[df4[5] == 25+(25*x)]
    pdfPRIOAODV_EM.append(tmp[2].mean())
    delayPRIOAODV_EM.append(tmp[4].mean())
    

pdfTest = df5[2].to_numpy().reshape(75, 1)
delayTest = df5[4].to_numpy().reshape(75, 1)
```

```python
plt.plot(connections, pdfAODV, marker='x', label="AODV")
plt.plot(connections, pdfDSR, marker='x', label="DSR")
plt.plot(connections, pdfPRIOAODV, marker='o', label="PRIO-AODV")
plt.plot(connections, pdfPRIOAODV_EM, marker='o', label="PRIO-AODV-E")
plt.legend(loc='upper right')
plt.ylabel('Packet Delivery Fraction (%)')
plt.xlabel('Connections')
plt.xticks(connections)
plt.rcParams["figure.figsize"] = (7,3)
plt.show()
```

```python
plt.plot(connections, delayAODV, marker='x', label="AODV")
plt.plot(connections, delayDSR, marker='x', label="DSR")
plt.plot(connections, delayPRIOAODV, marker='o', label="PRIO-AODV")
plt.plot(connections, delayPRIOAODV_EM, marker='o', label="PRIO-AODV-E")
plt.legend(loc='upper left')
plt.ylabel('Delay')
plt.xlabel('Connections')
plt.xticks(connections)
plt.rcParams["figure.figsize"] = (7,3)
plt.show()
```

```python
m = np.arange(75).reshape(75, 1)
n = np.arange(0,75,1).reshape(75, 1)
plt.plot(m, pdfTest)
plt.ylabel('PDR')
plt.xlabel('Seeds - 1')
plt.xticks(n)
plt.rcParams["figure.figsize"] = (40,3)
plt.show()
```

```python
plt.plot(m, delayTest)
plt.ylabel('Delay')
plt.xlabel('Seeds - 1')
plt.xticks(n)
plt.rcParams["figure.figsize"] = (40,3)
plt.show()
```
